# Preparación del Entorno

En este notebook se realizarán las siguientes tareas

1. Creación de un catálogo para el lakehouse
2. Creación de las external locations para landing y gold
3. Creación de los esquemas para landing, bronze, silver y gold
4. Creación del volumen de landing

```sql
CREATE CATALOG IF NOT EXISTS ecom_lakehouse
COMMENT 'Catálogo para el lakehouse';
```

```sql
CREATE EXTERNAL LOCATION landing_ext_loc
URL 'abfss://lakehouse@ecomlakehousesg.dfs.core.windows.net/files/landing/'
WITH STORAGE CREDENTIAL ecom_pipeline;

CREATE EXTERNAL LOCATION gold_ext_loc
URL 'abfss://lakehouse@ecomlakehousesg.dfs.core.windows.net/gold/'
WITH STORAGE CREDENTIAL ecom_pipeline;
```

```sql
CREATE SCHEMA IF NOT EXISTS ecom_lakehouse.files
COMMENT 'Schema para files'
LOCATION 'abfss://lakehouse@ecomlakehousesg.dfs.core.windows.net/files/';

CREATE SCHEMA IF NOT EXISTS ecom_lakehouse.bronze
COMMENT 'Schema para bronze'
MANAGED LOCATION 'abfss://lakehouse@ecomlakehousesg.dfs.core.windows.net/bronze/';

CREATE SCHEMA IF NOT EXISTS ecom_lakehouse.silver
COMMENT 'Schema para silver'
MANAGED LOCATION 'abfss://lakehouse@ecomlakehousesg.dfs.core.windows.net/silver/';

CREATE SCHEMA IF NOT EXISTS ecom_lakehouse.gold
COMMENT 'Schema para gold'
LOCATION 'abfss://lakehouse@ecomlakehousesg.dfs.core.windows.net/gold/'
```

```sql
CREATE EXTERNAL VOLUME IF NOT EXISTS ecom_lakehouse.files.landing
COMMENT 'Volumen para landing'
LOCATION 'abfss://lakehouse@ecomlakehousesg.dfs.core.windows.net/files/landing/'
```
